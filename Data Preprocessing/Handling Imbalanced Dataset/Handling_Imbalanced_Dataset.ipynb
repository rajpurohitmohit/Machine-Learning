## **Imbalanced Dataset**
A dataset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
credit_card_data.head()
```

```python
# Distribution of the two Classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 --> Legit Transactions  
1 --> Fraudulent Transactions

```python
# Seperating the Legit and Fraudulent Transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

### **Undersampling**
Creating a sample dataset containing similar distribution of Legit and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
# Sample = Give random values from the legit dataframe according to the n
legit_sample = legit.sample(n = 492)
```

```python
legit_sample.shape
```

#### **Concatenate the Two DataFrames**

```python
new_dataset = pd.concat([legit_sample, fraud], axis= 0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```
