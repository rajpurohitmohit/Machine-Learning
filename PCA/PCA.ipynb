```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
pd.set_option("display.max_columns",None)
```

```python
data =pd.read_csv("PCA_Dataset.csv")
```

```python
data.head(1)
```

```python
data_encoded = pd.get_dummies(data.select_dtypes(include="object"),drop_first=True)
```

```python
data_encoded = data_encoded.astype("int64")
```

```python
data_encoded.head(1)
```

```python
drop_col = data.select_dtypes(include="object")
```

```python
data.drop(drop_col,axis=1,inplace=True)
```

```python
data= pd.concat([data_encoded,data],axis=1)
```

```python
data.info()
```

```python
s = StandardScaler()
```

```python
data = s.fit_transform(data)
```

```python
pca = PCA(n_components=0.95)
```

```python
data = pca.fit_transform(data)
```
